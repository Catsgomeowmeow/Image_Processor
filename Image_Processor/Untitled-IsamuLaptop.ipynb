```python
#Loads the image file
image = input("file name (no extension): ")
if image == "":
    sys.exit("Please give a name")

#Converts the image to black and white (python uses a  high contrast version)
img_src = Image.open(image +'.jpg').convert('L')
#turns image into an array
im = np.array(img_src)
# print (im.shape)
sizeim = im.shape[0]
sizeim = sizeim /2
sizeim = int(sizeim)


data = im[sizeim,:]
y,x,_ = plt.hist(data)
print (x.max(),x.min())
gmax = x.max()
gmin = x.min()

#to get best image take the first peak of the mininmum 

```
